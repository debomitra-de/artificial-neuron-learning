```python
import numpy as np
import pandas as pd
data=pd.read_csv("mobile price train.csv")
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x.head(10)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)

model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test).round(2)*100)
```

```python
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
score=[]
k_range=range(1,31)

for k in k_range:
    model=KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    score.append(accuracy_score(y_pred,y_test).round(2)*100)
    
from matplotlib import pyplot as plt
plt.plot(k_range,score)
plt.xlabel('Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
print("first max of accuracy is k=",score.index(max(score))+1,"with accuracy of",max(score))
```
